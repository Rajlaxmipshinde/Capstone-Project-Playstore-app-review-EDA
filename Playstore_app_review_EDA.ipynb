

*  The 'price' column has '$' symbol. we need to drop it and change its datatype from object to float.



````python
playstore_df['Price'].unique()
````

````python
# Creating a function drop_dollar, which drops the $ symbol if present and returns the value as a float.
def drop_dollar(value):

    if '$' in value:
        return float(value[1:])
    else:
        return float(value)

# Applying the drop_dollar function to the 'Price' column
playstore_df['Price'] = playstore_df['Price'].apply(lambda x: drop_dollar(x))
````

````python
playstore_df['Price'].unique()
````



*  Remove the '+' symbol from 'Installs' column and change its datatype to int.



````python
playstore_df['Installs'].value_counts()
````

````python
def drop_plus(value):
    '''
    This function drops the + symbol if present and returns the value with int datatype.
    If the value is not a valid integer, return 0.
    '''
    try:
        if '+' and ',' in value:
            return int(value[:-1].replace(',', ''))
        elif '+' in value:
            return int(value[:-1])
        else:
            return int(value)
    except ValueError:
        return 0

# Applying the drop_plus function to the 'Installs' column
playstore_df['Installs'] = playstore_df['Installs'].apply(lambda x: drop_plus(x))
````

````python
playstore_df['Installs'].value_counts()
````



*   Convert the 'Size' column to a numerical format in megabytes (MB).



````python
#Function to convert the 'Size' column to a numerical format in megabytes (MB).
def convert_size_to_numeric(size):

    if size == 'Varies with device':
        return None  # Or any other value you prefer for such cases
    elif size.endswith('M'):
        return float(size[:-1])  # Already in MB
    elif size.endswith('k') or size.endswith('K'):
        return float(size[:-1]) / 1024  # Convert KB to MB
    else:
        return None  # Or handle other cases as needed

# Applying the convert_size_to_numeric function to the 'Size' column
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: convert_size_to_numeric(x))
````

````python
# Extract non-float values in 'Size' column
non_float_size_values = playstore_df['Size'][~playstore_df['Size'].apply(lambda x: isinstance(x, float))]

# Calculate the percentage of non-float values in 'Size' column
percentage_non_float = (len(non_float_size_values) / len(playstore_df['Size'])) * 100

# Print the result
print(f"Non-float values in the 'Size' column: {non_float_size_values.unique()}")
print(f"Percentage of non-float values in the 'Size' column: {percentage_non_float:.2f}%")

````

#### Chart - 2 : Distribution of Paid and Free Apps/Percentage of Paid and Free app installs

````python
# Chart - 2 visualization code
# Calculate the count of paid and free apps
paid_apps_count = playstore_df[playstore_df['Type'] == 'Paid'].shape[0]
free_apps_count = playstore_df[playstore_df['Type'] == 'Free'].shape[0]

# Create data for the pie chart (free_paid_install)
free_paid_install = playstore_df.groupby(['Type'])['Installs'].sum()

# Create data for the second pie chart (distribution of Paid and Free Apps)
labels_dist = ['Paid Apps', 'Free Apps']
sizes_dist = [paid_apps_count, free_apps_count]
colors_dist = ['lightblue', 'gold']
explode_dist = (0.1, 0)  # explode the first slice (Paid Apps) in the second pie chart

# Plotting both pie charts in a single figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

# First pie chart: Distribution of Paid and Free Apps
ax1.pie(sizes_dist, explode=explode_dist, labels=labels_dist, colors=colors_dist, autopct='%.2f%%', shadow=True, startangle=140)
ax1.set_title('Distribution of Paid and Free Apps')

# Second pie chart: Percentage of Paid and Free app installs
ax2.pie(free_paid_install, labels=free_paid_install.index,colors=colors_dist, autopct='%.2f%%',startangle=140, textprops={'fontsize': 10}, shadow=True)
ax2.set_title('Percentage of Paid and Free app installs')

# Equal aspect ratio ensures that pie charts are drawn as circles.
ax1.axis('equal')
ax2.axis('equal')

plt.tight_layout()
plt.show()
````

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code

# Select only the numerical columns
numerical_cols = ['Rating', 'Reviews', 'Size', 'Installs', 'Price']

# Compute the correlation matrix
corr_data = playstore_df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_data, cmap='PuRd', annot=True, fmt='.2f')
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
````

##### 1. Why did you pick the specific chart?

Rather than creating a heatmap of all columns in a Play Store dataset, it's more common and practical to generate a heatmap of the correlation matrix for numerical variables. This helps identify the most correlated variables and avoids overwhelming complexity.

##### 2. What is/are the insight(s) found from the chart?

1. The heatmap reveals a strong positive link between the number of reviews and the number of installs, with a correlation coefficient of roughly 0.63. According to this, apps that receive more reviews also frequently receive more installations.

2. Size and other factors do not strongly correlate: With correlation values ranging from -0.02 to 0.18, the heatmap reveals that the Size of the app has a poor link with other factors like Rating, Reviews, Installs, and Price.
